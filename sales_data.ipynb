```python
top_products = sales.sort_values(by='Total_Revenue', ascending=False).head(3)
top_products
```

```python
revenue_bar = plt.bar(sales['Product_Category'], sales['Total_Revenue'])
revenue_bar
```

```python
revenue_by_region = sales.groupby('Region')['Total_Revenue'].sum()
revenue_by_region
plt.pie(revenue_by_region, labels=revenue_by_region.index, autopct = '%1.1f%%', startangle=90)
plt.title('Revenue by Region')
```
